```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
records = pd.read_csv('results.csv')
```

```python
plt.subplots(figsize=(20, 10))
a = sb.boxplot(data = records, x = 'Hour_Of_Day', y='High_Stock_Price')
plt.title("Price Range of All Stocks by Hour", fontsize=15)
plt.xlabel("Hour_Of_Day")
plt.ylabel("High_Stock_Price")
```

```python
# records["Change"] = records['High_Stock_Price'].pct_change()
# df = records.dropna(axis=0, how='any').round(2)

BYND = records.loc[records['Company_Name'] == 'BYND'].round(2)
DDOG = records.loc[records['Company_Name'] == 'DDOG'].round(2)
FB = records.loc[records['Company_Name'] == 'DDOG'].round(2)
NFLX = records.loc[records['Company_Name'] == 'NFLX'].round(2)
OKTA = records.loc[records['Company_Name'] == 'OKTA'].round(2)
PINS = records.loc[records['Company_Name'] == 'PINS'].round(2)
SHOP = records.loc[records['Company_Name'] == 'SHOP'].round(2)
SNAP = records.loc[records['Company_Name'] == 'SNAP'].round(2)
SQ = records.loc[records['Company_Name'] == 'SQ'].round(2)
TTD = records.loc[records['Company_Name'] == 'TTD'].round(2)
```

```python
BYND['Change'] = BYND['High_Stock_Price'].pct_change()
DDOG['Change'] = DDOG['High_Stock_Price'].pct_change()
FB['Change'] = FB['High_Stock_Price'].pct_change()
NFLX['Change'] = NFLX['High_Stock_Price'].pct_change()
OKTA['Change'] = OKTA['High_Stock_Price'].pct_change()
PINS['Change'] = PINS['High_Stock_Price'].pct_change()
SHOP['Change'] = SHOP['High_Stock_Price'].pct_change()
SNAP['Change'] = SNAP['High_Stock_Price'].pct_change()
SQ['Change'] = SQ['High_Stock_Price'].pct_change()
TTD['Change'] = TTD['High_Stock_Price'].pct_change()
df = pd.concat([BYND,DDOG,FB,NFLX,OKTA,SHOP,PINS,SNAP,SQ,TTD]).dropna(axis=0, how='any')
```

```python
plt.subplots(figsize=(20, 10))
plt.title('Percent Change From Highest Price Per Hour', fontsize=15)
sb.lineplot(data=df, x='Hour_Of_Day',y='Change',hue='Company_Name')
```

```python
plt.subplots(figsize=(20, 10))
plt.title('Highest Stock Price Hourly Fluctuations per Stock', fontsize=15)
sb.lineplot(data=records, x='Hour_Of_Day',y='High_Stock_Price',hue='Company_Name')
```
